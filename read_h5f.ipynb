```python
import os
import h5py
import numpy as np
```

```python
# open the HDF5 file specified by the path
pid = "20250305.50"
# build h5py_path with 3-digit zero-padded fractional part after the dot
head, sep, tail = pid.strip().rpartition('.')
if sep:
    frac = tail
    if frac.isdigit() and len(frac) < 3:
        frac = frac.zfill(3)
    pid_formatted = f"{head}_{frac}"
else:
    pid_formatted = pid.replace('.', '_')

h5py_path = f"/home/IPP-HGW/orluca/devel/data/HDF/_data/{pid_formatted}/"
data_dict = {} # dictionary to store the diodes
for f in os.listdir(h5py_path):
    # skip the first four files
    if f.endswith("_t000.h5f") or f.endswith("_t090.h5f") or f.endswith("_t180.h5f") or f.endswith("_t270.h5f"):
        h5f_file = os.path.join(h5py_path, f)
        with h5py.File(h5f_file, "r") as h5f:
            # read the XMCTSdata group
            if "XMCTSdata" in h5f:
                xmcts_data = h5f["XMCTSdata"]
                # read all of the datasets in the XMCTSdata group but keep only the diodes
                for dataset in xmcts_data.keys():
                    if dataset.endswith("0"):
                        data_dict["time_base"] = xmcts_data[dataset][:]
        continue # skip the first four files and continue to the next iteration
    if f.endswith(".h5f"):
        h5f_file = os.path.join(h5py_path, f)
        with h5py.File(h5f_file, "r") as h5f:
            if "XMCTSdata" in h5f:
                xmcts_data = h5f["XMCTSdata"]
                # read all of the datasets in the XMCTSdata group
                for dataset in xmcts_data.keys():
                    if not dataset == "timedata":
                        data_dict[dataset] = xmcts_data[dataset][:]
    
```
